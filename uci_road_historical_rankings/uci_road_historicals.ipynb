```
# Process for loading historical uci rankings

# Import pandas
import pandas as pd

# set new root path
root_path = '/content/drive/My Drive/cycling_data/Cycling_Yearly_data'

# import data sets
df1 = pd.read_excel('/content/drive/My Drive/cycling_data/Cycling_Yearly_data/IndividualRanking_2019.xlsx')
df2 = pd.read_excel('/content/drive/My Drive/cycling_data/Cycling_Yearly_data/IndividualRanking_2018.xlsx')
df3 = pd.read_excel('/content/drive/My Drive/cycling_data/Cycling_Yearly_data/IndividualRanking_2017.xlsx')
df4 = pd.read_excel('/content/drive/My Drive/cycling_data/Cycling_Yearly_data/IndividualRanking_2016.xlsx')
df5 = pd.read_excel('/content/drive/My Drive/cycling_data/Cycling_Yearly_data/IndividualRanking_2015.xlsx')

# Add a year column for each dataset
df1['year'] = 2019
df2['year'] = 2018
df3['year'] = 2017
df4['year'] = 2016
df5['year'] = 2015

# Add a year column for each dataset
df1['type'] = 'road_wt'
df2['type'] = 'road_wt'
df3['type'] = 'road_wt'
df4['type'] = 'road_wt'
df5['type'] = 'road_wt'

# append all the data sets to one
df1 = df1.append(df2, ignore_index = True)
df1 = df1.append(df3, ignore_index = True)
df1 = df1.append(df4, ignore_index = True)
df1 = df1.append(df5, ignore_index = True)

# save the master file to a csv
df1.to_csv('/content/drive/My Drive/cycling_data/uci_historical_rankings/uci_road_rankins_historicals.csv', index = False)

```
